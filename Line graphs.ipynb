```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
a = np.array([1,2,3,4,5])
a

plt.plot(a)
plt.show()
```

```python
#creating the raw data
a = np.array([1,2,3,4,5])
b = np.array([2,2,2,2,2])

#inreasing size
plt.figure(figsize=(3,5))

plt.plot(a , c = 'green')
plt.plot(b , c = 'pink')
plt.show()
```

```python
#creating the raw data 
a = np.array([1,2,3,4,5])
b = np.array([2,2,2,2,2])
d = np.array([1,7,9,4,2])
#Line width and changing line style by making it dotted
plt.figure(figsize=(3,5))

plt.plot(a , c = 'green' , linestyle = '-.')
plt.plot(b , linewidth = 5.5)
plt.plot(d , c = 'black' , linestyle = '--' )
plt.title("Rent per month Summary")
plt.xticks(range(0,len(b)) , ('Jan' , 'Feb' , 'March' , 'April' , 'May'))
plt.show()

```
